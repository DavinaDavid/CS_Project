```python
#1. Create a single dataframe with the concatenation of all input csv files, adding a column called country
```

```python
my_df_files = [CA_Videos,DE_Videos,FR_Videos,GB_Videos,IN_Videos,JP_Videos, KR_Videos, MX_Videos,RU_Videos,US_Videos]
my_df_files

# Corresponding country names
countries = ['Canada','Germany','France','Great Britain','India','Japan','Korea', 'Mexico', 'Russia', 'United States']

new_df = []

for df, country in zip(my_df_files, countries):
    df_copy = df.copy()       # Optional: make a copy to avoid changing original
    df_copy['Country'] = country
    new_df.append(df_copy)

# Concatenate all DataFrames
All_YT_Videos = pd.concat(new_df, ignore_index=True)

All_YT_Videos.head()
```

```python
All_YT_Videos.info()
```

```python
#2. Extract all videos that have no tag.
```

```python
NO_tag = All_YT_Videos[All_YT_Videos['tags']=='[none]']
NO_tag
```

```python
#3. For each channel, determine the total number of views
```

```python
All_YT_Videos.groupby('channel_title')['views'].sum()
```

```python
#4. Save all rows with disabled comments and disabled ratings, or that have video_error_or_removed in a new dataframe called excluded,
#and remove those rows from the original dataframe.
```

```python
excluded = All_YT_Videos.loc[(All_YT_Videos['comments_disabled']==True) | (All_YT_Videos['ratings_disabled']==True) | (All_YT_Videos['video_error_or_removed']==True)]
#excluded
#remove from original df
All_YT_Videos.drop(excluded.index, inplace=True)
All_YT_Videos
```

```python
#5. Add a like_ratio column storing the ratio between the number of likes and of dislikes
```

```python
import numpy as np
All_YT_Videos['like_ratio'] = All_YT_Videos['likes']/All_YT_Videos['dislikes'].replace(0,np.nan) #incase divided by 0 replace to nan
All_YT_Videos['like_ratio']
```

```python
#6.Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)
```

```python
All_YT_Videos['publish_time'] = pd.to_datetime(All_YT_Videos['publish_time'])
#All_YT_Videos['publish_time']
All_YT_Videos['publish_time_10Min'] = All_YT_Videos['publish_time'].dt.floor('10min')
#All_YT_Videos['publish_time_10Min']
All_YT_Videos.groupby('publish_time_10Min')['video_id'].count()
```
