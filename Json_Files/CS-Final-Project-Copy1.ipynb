```python
#8 For each tag, determine the number of videos
```

```python
#Copy Dataset to avoid Original data loss
Copied_All_YT_Videos = All_YT_Videos.copy()
Copied_All_YT_Videos['tags_split'] = Copied_All_YT_Videos['tags'].str.split('|')
#All_YT_Videos['tags_split']

#Create own row for each tag
exploded_tags = Copied_All_YT_Videos.explode('tags_split')

#-----Data Cleaning----
#Remove "" and extra white spaces
exploded_tags['tags_split'] = exploded_tags['tags_split'].str.replace('"', '').str.strip()
#Remove rows where NA or missing
exploded_tags = exploded_tags[exploded_tags['tags_split'].notna()]
#Renaming
exploded_tags['tags_split'] = exploded_tags['tags_split'].replace({'[none]':'No Tags', '':'No Tags'})
exploded_tags['tags_split']

#Count tags
tag_counts = exploded_tags['tags_split'].value_counts()
tag_counts
```

```python
#9. Find the tags with the largest number of videos
```

```python
#Remove count of No-tag
tag_counts_filtered = tag_counts[tag_counts.index != 'No Tags']
#find Max Count
max_tag_count = tag_counts_filtered.max()
#max_tag_count #returns count but not the tag name
top_video_tag = tag_counts_filtered[tag_counts_filtered == max_tag_count]
top_video_tag
```

```python
#14. Read all json files with the video categories
```

JSON file reading help from Chatgpt --Davina

```python
import json
import glob
import os

category_frames = []

#for each file in Json_files
for file in glob.glob("Json_Files/*_category_id.json"):
   #Open files in read-mode
    with open(file, "r") as f: 
        data = json.load(f) 

    # Normalize JSON structuring
    df_json_structuring = pd.json_normalize(data["items"])

    # Add a country column (CA, FR, etc.)
    country = os.path.basename(file).split("_")[0]
    df_json_structuring["Country"] = country

    category_frames.append(df_json_structuring)

    # Combine all JSON category data into one DataFrame
categories = pd.concat(category_frames, ignore_index=True)
categories.head()
```

```python
#Converting catergory id To int as found in ALL_YT_Videos to apply merge
categories['id'] = categories['id'].astype(int)

#Applying Left Join to get common rows of All_YT_Videos
merged_table = All_YT_Videos.merge(categories, left_on=['category_id', 'Country'],
                             right_on=['id', 'Country'], how='left')
#Filtering
#Note: NA values are also unassignable
merged_table['snippet.assignable'] = merged_table['snippet.assignable'].astype('boolean') #to resolve panda future version warning, converting it first to bool
non_assignable_videos = merged_table[merged_table['snippet.assignable'].fillna(False) == False]

result_non_assignable_videos = non_assignable_videos.groupby('Country')['video_id'].count().reset_index()
result_non_assignable_videos_renamed = result_non_assignable_videos.rename(columns={'video_id': 'non_assignable_videos_count'})
result_non_assignable_videos_renamed
```
